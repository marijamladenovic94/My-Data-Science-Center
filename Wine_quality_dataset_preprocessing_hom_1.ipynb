### Dealing with missing data

```python
# Missing values in whole dataset
data.isnull().sum()
```

```python
# Display all rows with any nan value
data[data.isnull().any(axis=1)]
```

As we can see from previous table we have a small number of missing values of some columns. We will try to solve this, but we don't drop all columns with missing values because we will lose a lot of data. We are going to use different approaches to deal with missing values.

```python
# First approach of dealing with missing values is to drop them
data_dropna = data.dropna()
data_dropna
```

```python
# Second approach is to fill missing values with mean, median, mode, etc
from sklearn.impute import SimpleImputer

imr = SimpleImputer(missing_values=np.nan, strategy='mean')
imr = imr.fit(X = data.drop('type', axis = 1))
imputed_data = imr.transform(X = data.drop('type', axis = 1).values)
imputed_data
```

```python
imputed_data = pd.DataFrame(data = imputed_data, columns = list(data.columns)[1:])
imputed_data.isnull().sum()
```

```python
imputed_data_df = pd.merge(data['type'], imputed_data, left_index=True, right_index=True)
imputed_data_df.head()
```

### Split dataset on train and test

```python
X = imputed_data_df.drop('quality', axis = 1)
y = imputed_data_df.quality
print(y)
X.head()
```

```python
# Split independent and target variable on train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

### Handling categorical data

```python
# First we need to see how much unique values we have in column type
X_train['type'].unique()
```

```python
# Transform train and test dataset using LabelEncoder because we have just 2 different values
from sklearn.preprocessing import LabelEncoder

class_le = LabelEncoder()
X_train['type'] = class_le.fit_transform(X_train.loc[: , 'type'])
display(X_train)

X_test['type'] = class_le.fit_transform(X_test.loc[: , 'type'])
X_test['type']
```

### Bringing features onto the same scale

```python
# First we will try to normalize data and than we will observe the distribution of data
from sklearn.preprocessing import MinMaxScaler

# fit scaler on training data
norm = MinMaxScaler().fit(X_train)

# transform training data
X_train_norm = norm.transform(X_train)

# transform testing dataabs
X_test_norm = norm.transform(X_test)
```

```python
# Create dataframe of training normalized data
X_train_norm_df = pd.DataFrame(data = X_train_norm, columns = list(X_train.columns))
```

```python
# Let see the outliers for each continue variable
sub_boxplot(X_train_norm_df)
```

```python
# Now let try to standardize train and test data
from sklearn.preprocessing import StandardScaler

stdsc = StandardScaler()
# transform training data
X_train_std = stdsc.fit_transform(X_train)
# transform testing data
X_test_std = stdsc.transform(X_test)
```

```python
# Create dataframe of training standardized data
X_train_std_df = pd.DataFrame(data = X_train_norm, columns = list(X_train.columns))
```

```python
# Let see the outliers for each continue variable
sub_boxplot(X_train_std_df)
```
