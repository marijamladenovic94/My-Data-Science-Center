```python
#import libraries
import os #provides functions for interacting with the operating system
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#import dataset for visualization
tips = sns.load_dataset("tips")
tips
```

### Description of each column

Total bill - cost of the meal, including tax, in US dollars

Tip - gratuity in US dollars

Sex - sex of person paying for the meal 

Smoker - smoker in party

Day - days per week

Time - time per day

Size - size of the party

### Distribution of continuous variables

```python
# total_bill
sns.distplot(tips['total_bill'].values, color = 'blue', bins = 15).set_title('Dist of total_bill')
```

```python
# tips
sns.distplot(tips['tip'].values, color = 'blue', bins = 15).set_title('Dist of tip')
```

```python
# Set up the matplotlib figure
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# plot 1

sns.boxplot(x="total_bill", y="sex", data=tips, color = '#52F954', ax=axes[0,0]).set_title('Dist of total_bill per sex')
sns.swarmplot(x="total_bill", y="sex", data=tips, color="blue", ax=axes[0,0])

# plot 2
sns.boxplot(x="total_bill", y="smoker", data=tips, color = '#52F954', ax=axes[0,1]).set_title('Dist of total_bill per smoking')
sns.swarmplot(x="total_bill", y="smoker", data=tips, color="blue", ax=axes[0,1])

# plot 3
sns.boxplot(x="total_bill", y="day", data=tips, color = '#52F954', ax=axes[1,0]).set_title('Dist of total_bill per day')
sns.swarmplot(x="total_bill", y="day", data=tips, color="blue", ax=axes[1,0])

# plot 4
sns.boxplot(x="total_bill", y="time", data=tips, color = '#52F954', ax=axes[1,1]).set_title('Dist of total_bill per time of day')
sns.swarmplot(x="total_bill", y="time", data=tips, color="blue", ax=axes[1,1])
```

```python
# Set up the matplotlib figure
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# plot 1

sns.boxplot(x="tip", y="sex", data=tips, color = '#52F954', ax=axes[0,0]).set_title('Dist of tip per sex')
sns.swarmplot(x="tip", y="sex", data=tips, color="blue", ax=axes[0,0])

# plot 2
sns.boxplot(x="tip", y="smoker", data=tips, color = '#52F954', ax=axes[0,1]).set_title('Dist of tip per smoking')
sns.swarmplot(x="tip", y="smoker", data=tips, color="blue", ax=axes[0,1])

# plot 3
sns.boxplot(x="tip", y="day", data=tips, color = '#52F954', ax=axes[1,0]).set_title('Dist of tip per day')
sns.swarmplot(x="tip", y="day", data=tips, color="blue", ax=axes[1,0])

# plot 4
sns.boxplot(x="tip", y="time", data=tips, color = '#52F954', ax=axes[1,1]).set_title('Dist of tip per time of day')
sns.swarmplot(x="tip", y="time", data=tips, color="blue", ax=axes[1,1])
```

```python
# Let see the correlation between total_bill and tip per day, time, sex, smooking
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# plot 1

sns.scatterplot(x="total_bill", y="tip", hue = 'day', data=tips, style = 'day', color = 'green', size = 'tip', ax=axes[0,0]).set_title('total_bill vs tip per day')

# plot 2
sns.scatterplot(x="total_bill", y="tip", hue = 'time', data=tips, style = 'time', size = 'tip', color = 'green', ax=axes[0,1]).set_title('total_bill vs tip per time')

# plot 3
sns.scatterplot(x="total_bill", y="tip", hue = 'sex', data=tips, style = 'sex', size = 'tip', color = 'green', ax=axes[1,0]).set_title('total_bill vs tip per sex')

# plot 4
sns.scatterplot(x="total_bill", y="tip", hue = 'smoker', data=tips, style = 'smoker', size = 'tip', color = 'green', ax=axes[1,1]).set_title('total_bill vs tip per smooking')
```

```python
# Let add two more columns total_bill per 1 person and tips per 1 person
tips_2 = tips.copy()
tips_2.head()
```

```python
tips_2['total_bill_person'] = tips_2['total_bill']/tips_2['size']
tips_2['tip_person'] = tips_2['tip']/tips_2['size']
tips_2.head()
```

```python
# Let see the correlation between total_bill and tip per day, time, sex, smooking
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# plot 1

sns.scatterplot(x="total_bill_person", y="tip_person", hue = 'day', data=tips_2, style = 'day', color = 'green', size = 'tip', ax=axes[0,0]).set_title('total_bill vs tip per day')

# plot 2
sns.scatterplot(x="total_bill_person", y="tip_person", hue = 'time', data=tips_2, style = 'time', size = 'tip', color = 'green', ax=axes[0,1]).set_title('total_bill vs tip per time')

# plot 3
sns.scatterplot(x="total_bill_person", y="tip_person", hue = 'sex', data=tips_2, style = 'sex', size = 'tip', color = 'green', ax=axes[1,0]).set_title('total_bill vs tip per sex')

# plot 4
sns.scatterplot(x="total_bill_person", y="tip_person", hue = 'smoker', data=tips_2, style = 'smoker', size = 'tip', color = 'green', ax=axes[1,1]).set_title('total_bill vs tip per smooking')
```

```python
# Correlation - heatmap
pc = tips[['total_bill', 'tip', 'size']].corr(method ='pearson')
ax = sns.heatmap(pc, annot=True,
                 yticklabels=['total_bill', 'tip', 'size'],
                 xticklabels=['total_bill', 'tip', 'size'],
                 annot_kws={'size': 20},
                linecolor='blue', center=False)
```
