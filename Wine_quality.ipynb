```python
# Let create function for subplots of boxplots
def sub_boxplot(data, num_col = None, y=None, hue=None, orient='h', *args):
    if data.shape[1] %2 == 0:
        i = data.shape[1]//num_col
    else:
        i = data.shape[1]//num_col + 1
    fig, axes = plt.subplots(i, num_col, figsize=(15, 25))
    for i in range(i):
        for j in range(num_col):
            if (2*i+j) < data.shape[1]:
                sns.boxplot(x = data.columns[2*i+j], y=y, data=data, hue=hue, orient=orient , ax=axes[i,j] )
            else:
                axes[i,j].set_axis_off()
```

#### Binning target variable

Since we have a little number of wines with quality 3,4 we will merge tham with qulity 5 and this will be poor quality. The same we will do with quality 8 and 9 and we merge them with quality 7 which will be good quality. And the quality 6 will be middle quality. So, we will do binning: poor quality 0, middle quality 1 and good quality 2.

```python
# First binning
data.loc[data['quality'] == 6, 'quality_binn'] = 1
data.loc[(data['quality'] < 6) , 'quality_binn'] = 0
data.loc[data['quality'] > 6, 'quality_binn'] = 2 
# Second binning
data.loc[(data['quality'] < 6) , 'quality_binn_2class'] = 0
data.loc[data['quality'] >= 6, 'quality_binn_2class'] = 1 
data.head()
```

```python
# Let see the distribution of the data
# First binning with 3 classes
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
sns.countplot(data['quality'], ax=axes[0] ).set_title('Dist of quality')
sns.countplot(data['quality_binn'], ax=axes[1] ).set_title('Dist of quality with 3 classes')
sns.countplot(data['quality_binn_2class'], ax=axes[2] ).set_title('Dist of quality with 2 classes')
```

### Dealing with missing data

```python
# Missing values in whole dataset
data.isnull().sum()
```

```python
# Display all rows with any nan value
data[data.isnull().any(axis=1)]
```

As we can see from previous table we have a small number of missing values of some columns. We will try to solve this, but we don't drop all columns with missing values because we will lose a lot of data. We are going to use different approaches to deal with missing values.

```python
# First approach of dealing with missing values is to drop them
data_dropna = data.dropna()
data_dropna
```

```python
# Second approach is to fill missing values with mean, median, mode, etc
from sklearn.impute import SimpleImputer

imr = SimpleImputer(missing_values=np.nan, strategy='mean')
imr = imr.fit(X = data.drop('type', axis = 1))
imputed_data = imr.transform(X = data.drop('type', axis = 1).values)
imputed_data
```

```python
imputed_data = pd.DataFrame(data = imputed_data, columns = list(data.columns)[1:])
imputed_data.isnull().sum()
```

```python
imputed_data_df = pd.merge(data['type'], imputed_data, left_index=True, right_index=True)
imputed_data_df.head()
```

### Split dataset on train and test

```python
X = imputed_data_df.drop(['quality', 'quality_binn', 'quality_binn_2class'], axis = 1)
y = imputed_data_df.quality_binn
print(y)
X.head()
```

```python
# Split independent and target variable on train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

#### SVC

```python
# First we will 
from sklearn.svm import SVC, NuSVC, LinearSVC
from sklearn.metrics import confusion_matrix 
def train_model(model, X, y, X_test, y_test):
    clf = model()
    clf.fit(X, y)
    y_pred = clf.predict(X_test)
    # model accuracy for X_test   
    accuracy = clf.score(X_test, y_test) 
  
    # creating a confusion matrix 
    cm = confusion_matrix(y_test, y_pred) 
    print(f"Accuracy: {accuracy} \n")
    print(f"Confusion matrix: \n {cm}")
```

```python
# Import cross_val_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
```

```python
# Create dict acc_scores to see which model is the best
acc_scores = {}
```

```python
# First model SVC kernel linear
from sklearn.metrics import accuracy_score
clf = SVC(kernel='linear', C=1)
scores = cross_val_score(clf, X_train_norm, y_train, cv=5)
print(scores)
print(scores.mean())
clf.fit(X_train_norm, y_train)
y_pred = clf.predict(X_test_norm)
acc_scores['SVC_linear'] = accuracy_score(y_test, y_pred)
```

```python
# Define function for train,fit and predict new outcomes 
def train_model(model, X, y, X_test, y_test, model_name):
    clf = model
    scores = cross_val_score(clf, X_train_norm, y_train, cv=5)
    print(f"Cross_val_scores: {scores} \n ")
    print(f"Cross_val_scores_mean: {scores.mean()} \n ")
    print(f"Cross_val_scores_std: {scores.std()} \n")
    clf.fit(X_train_norm, y_train)
    y_pred = clf.predict(X_test_norm)
    acc_scores[model_name] = accuracy_score(y_test, y_pred)
    print(f"Acc_score of {model_name}: {acc_scores[model_name]}")
    return model, acc_scores
```
